```python
#read errores.txt and print the lines that contain the word "error"
with open("errores.txt") as file:
    for line in file:
        if "error" in line.lower():
            print(line, end="")
```

```python
#crear un dataframe con los errores "Epoca 0 Error total: 1.035188"
import pandas as pd
import re

errors = []
with open("errores.txt") as file:
    for line in file:
        match = re.search(r"Epoca (\d+) Error total: (\d+\.\d+)", line)
        if match:
            errors.append((int(match.group(1)), float(match.group(2))))

df = pd.DataFrame(errors, columns=["Epoca", "Error"])
print(df)
```

```python
#graficar el error en funcin de la poca - idioma ingls
import matplotlib.pyplot as plt

plt.plot(df["Epoca"], df["Error"])
plt.xlabel("Epoch")
plt.ylabel("Error")
plt.title("Error vs Epoch")
plt.show()


```
